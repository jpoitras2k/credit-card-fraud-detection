# Phase 1: Data Analysis and Preparation

In this notebook, we perform Exploratory Data Analysis (EDA) on the Credit Card Fraud Detection dataset to understand the features, visualize data distributions, and check for class imbalances.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Adjust plotting style
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
import os
import sys

# Setup path to import our custom dataloader
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from ml_framework_project.datasets.credit_card import get_creditcard_data

df = get_creditcard_data()
df.head()
```

## Class Imbalance Visualization

Credit Card fraud datasets are typically highly imbalanced.

```python
counts = df['Class'].value_counts()
print(f"Genuine Transactions: {counts[0]}")
print(f"Fraudulent Transactions: {counts[1]}")
print(f"Percentage of Fraud: {counts[1] / (counts[0]+counts[1]) * 100:.3f}%")

plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Class')
plt.title('Distribution of Transaction Classes (0: Genuine, 1: Fraud)')
plt.show()
```

## Transaction Time and Amount Distributions

Since V1-V28 are PCA components, the only non-transformed features are `Time` and `Amount`.

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

sns.histplot(df['Amount'], bins=50, ax=ax[0], color='blue', kde=True)
ax[0].set_title('Transaction Amount Distribution')
ax[0].set_xlim([0, 2000]) # Zoomed in as most amounts are small

sns.histplot(df['Time'], bins=50, ax=ax[1], color='red', kde=True)
ax[1].set_title('Transaction Time Distribution')

plt.show()
```

## Correlation Matrix

Visualizing correlations helps us understand feature significance. (Note: PCA features V1-V28 will have 0 correlation with each other, but might correlate with Time/Amount/Class).

```python
plt.figure(figsize=(12, 10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm_r', annot_kws={'size':20})
plt.title('Feature Correlation Matrix')
plt.show()
```
